```python
import pandas as pd

# BASIC

data=['ABC','DEF','GHI']
ds=pd.Series(data)
print(ds)
print(type(ds))

# INDEX

ds1=pd.Series(data,index=['A','B','C'])
print(ds1)

# INDEX WITH VALUES

d={'A':20,'B':30,'C':40}
ds2=pd.Series(d)
print(ds2)

# RANGE

ds3=pd.Series(range(51,61))
print(ds3)
```

# DataFrame()


```python
import pandas as pd

data={'name':['Roy','Harsh','Om','Veer'],
      'div':['D1','D2','D3','D4'],
       'marks':[30,40,28,26]}
df=pd.DataFrame(data)
print(df)
print(type(df))

# list in list

d=[['A',20,'D1'],
   ['A',20,'D1'],
   ['A',20,'D1']]
df1=pd.DataFrame(d,columns=['name','marks','div'],index=['AA','BB','CC'])
print(df1)

# LIST OF DICT

s1=pd.Series([20,21,22])
s2=pd.Series([23,24,25])
s3=pd.Series([26,27,28])
df2=pd.DataFrame({'col 1':s1,'col 2':s2})
print(df2)

# DICT in LIST

d1=[{'A':20,'B':30},
    {'A':21,'B':40},
    {'A':22,'B':50}]
df3=pd.DataFrame(d1)
print(df3)
```
